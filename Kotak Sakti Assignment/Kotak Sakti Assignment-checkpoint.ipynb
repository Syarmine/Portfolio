## Exploratory Data  Analysis

Exploratory Data Analysis (EDA) is a crucial step in any data analysis project as it helps us to understand the structure of the data, uncover patterns and relationships, and identify potential outliers or anomalies. The goal of EDA is to gain insights into the data that can help inform business decisions.

In this section, we will answer the Kotak Sakti business's questions, specifiying datasets combining SQL queries with filter functions.

```python
# Question 2b: number of customers purchasing more than 5 books
q_1 = """SELECT COUNT(quantity) AS customers_with_over_5_purchases
            FROM bookstore.invoiceslines
            WHERE quantity >5
            ;
            """

# Question 2c: customes who never purchased anything
q_2 = """SELECT *
            FROM customers
            WHERE customer_id NOT IN (SELECT DISTINCT invoices.customer_id
                                      FROM invoices)
            """
# Question 2d: Show the SQL query for list of book purchased with the users
q_3 = """SELECT DISTINCT invoiceslines.invoices_description, invoices.customer_id, customers.first_name
FROM invoices
INNER JOIN customers ON invoices.customer_id = customers.customer_id
INNER JOIN invoiceslines ON invoices.invoices_id = invoiceslines.invoices_id"""
```

```python
# Assign variable names of each data frame
# Question 2b: number of customers purchasing more than 5 books
df_2b = pd.read_sql(q_1, conn)

# Question 2c: customes who never purchased anything
df_2c = pd.read_sql(q_2, conn)

# Question 2d: Show the SQL query for list of book purchased with the users
df_2d = pd.read_sql(q_3, conn)
```
